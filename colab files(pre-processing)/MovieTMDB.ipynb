```python
temp = top.apply(lambda x: pd.Series(x['genres']),axis=1).stack().reset_index(level=1, drop=True)
temp.name = 'genre'
newtop = top.drop('genres', axis=1).join(temp)
```

```python
def make_toplist(genre):
    df = newtop[newtop['genre'] == genre]
    qualified = df.sort_values('popularity', ascending=False).head(250)
    
    return qualified
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i)
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
movies['actor_1'] = movies['cast'].map(lambda x: get_actor1(x))
```

```python
movies['actor_1']
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i)
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
movies['actor_2'] = movies['cast'].map(lambda x: get_actor2(x))
```

```python
movies['actor_2']
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i)
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
movies['actor_3'] = movies['cast'].map(lambda x: get_actor3(x))
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmitizer=WordNetLemmatizer()
```

```python
def lem(text):
  y=[]

  for i in text.split():
    y.append(lemmitizer.lemmatize(i))

  return " ".join(y)
```

```python
lem(new['tags'][4])
```

```python
new['tags']=new['tags'].apply(lem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```
